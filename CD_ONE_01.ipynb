# Conhecendo a base de dados

A base de dados diz respeito sobre o valor de imóveis oferecendo features como:
1. Tipo

## Importando os dados


```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep = ';')
dados.head()
```

## Características gerais da base de dados

```python
def caracteristicas (dataset):
    print(f'Tamanho: {dataset.shape}\n')
    print(f'Colunas: {dataset.columns}\n')
    print(f'Informações: {dataset.info()}\n')
    print(f'Descrição: {dataset.describe()}\n')
    print(f'Valores nulos: {dataset.isnull().sum()}')
```

```python
caracteristicas(dados)
```

# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/millenagena/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com '  + \
                     dados['Quartos'].astype(str) + ' quarto(s) '  + \
                     ' e '  + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
